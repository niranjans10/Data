```python
data["show_name"] = data["show_name"].fillna(data["show_name"].mode())
```

```python
data["aired_on"] = data["aired_on"].fillna(data["aired_on"].mode())
```

```python
data["original_network"] = data["original_network"].fillna(data["original_network"].mode())
```

```python
data["rating"] = data["rating"].fillna(data["rating"].mean())
```

```python
data["current_overall_rank"] = data["current_overall_rank"].fillna(data["current_overall_rank"].mean())
```

```python
data["watchers"] = data["watchers"].fillna(data["watchers"].mean())
```
